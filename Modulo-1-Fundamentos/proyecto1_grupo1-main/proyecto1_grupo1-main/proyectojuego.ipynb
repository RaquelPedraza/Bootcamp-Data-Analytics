## Juego 1. Piedra, Papel o Tijeras

Piedra Papel o Tijera: este es un juego clÃ¡sico en el que dos jugadores seleccionan una de tres opciones: Piedra, Papel o Tijera. Cada opciÃ³n tiene una relaciÃ³n de victoria diferente: la Piedra vence a la Tijera, la Tijera vence al Papel y el Papel vence a la Piedra. El objetivo del juego es elegir la opciÃ³n que vencerÃ¡ a la opciÃ³n seleccionada por el oponente. El juego se juega en rondas y el jugador que gana mÃ¡s rondas es el ganador del juego.

### CÃ³digo

```python
# Juego 1. Piedra, papel o tijera

# Para generar nÃºmeros aleatorios se hace uso del mÃ³dulo random
import random

# Se personaliza el juego pidiendo el nombre
nombre = input("Â¿CÃ³mo te llamas?").title()

# Introducimos las reglas del juego
print(f"Â¿Preparado para jugar a Piedra ğŸ‘Š, papel ğŸ– o tijera âœŒ, {nombre}?\n")
print(f"Reglas:\n - Piedra ğŸ‘Š gana a tijera âœŒ\n - Tijera âœŒ gana a papel ğŸ–\n - Papel ğŸ– gana a piedra ğŸ‘Š\n - El primero que llegue a 3 puntos gana.\n")
print("Elige una opciÃ³n.\n")

# Se han introducido las opciones en una tupla y utilizamos el codigo random para "una elecciÃ³n aleatoria"
opciones = ("piedra", "papel", "tijera")

# Se inicia el juego
jugar = "s"

while jugar == "s":

    # Se guarda el puntaje en variables y se contabilizan las rondas
    puntos_jugador = 0
    puntos_computadora = 0
    ronda = 0

    # Mientras el jugador y el computadora no lleguen a 3 puntos se sigue jugando
    while puntos_jugador < 3 and puntos_computadora < 3:

        # Eleccion aleatoria de la computadoraion escogida por el computadora 
        computadora = random.choice(opciones)

        # Se pide al jugador que escoja entre las 3 opciones
        jugador=str(input("Piedra ğŸ‘Š, papel ğŸ– o tijera âœŒ")).lower()

        # Indico la ronda de juego
        ronda +=1
        print(f"Ronda {ronda}")

        # Primer posible resultado: empate
        if computadora==jugador:
            if computadora=="piedra":
                print(f"Â¡EMPATE! {nombre}: {jugador} ğŸ‘Š / Computadora: {computadora} ğŸ‘Š\n")
            elif computadora=="papel":
                print(f"Â¡EMPATE! {nombre}: {jugador} ğŸ– / Computadora: {computadora} ğŸ–\n")
            else:
                print(f"Â¡EMPATE! {nombre}: {jugador} âœŒ / Computadora: {computadora} âœŒ\n")
        # Posibles resultados
        elif computadora=="piedra" and jugador == "papel":
            # Cada victoria suma un punto para jugador o computadora
            puntos_jugador += 1 
            print(f"Â¡Has ganado esta ronda!\n{nombre}: {jugador} ğŸ– / Computadora: {computadora} ğŸ‘Š\nPuntos jugador/a: {puntos_jugador} / Puntos computadora: {puntos_computadora}\n")
        elif computadora=="piedra" and jugador == "tijera":
            puntos_computadora += 1
            print(f"Â¡Has perdido esta ronda!\n{nombre}: {jugador} âœŒ / Computadora: {computadora} ğŸ‘Š\nPuntos jugador/a: {puntos_jugador} / Puntos computadora: {puntos_computadora}\n")
        elif computadora=="tijera" and jugador == "papel":
            puntos_computadora += 1
            print(f"Â¡Has perdido esta ronda!\n{nombre}: {jugador} ğŸ– / Computadora: {computadora} âœŒ\nPuntos jugador/a: {puntos_jugador} / Puntos computadora: {puntos_computadora}\n")
        elif computadora=="tijera" and jugador == "piedra":
            puntos_jugador += 1
            print(f"Â¡Has ganado esta ronda!\n{nombre}: {jugador} ğŸ‘Š / Computadora: {computadora} âœŒ\nPuntos jugador/a: {puntos_jugador} / Puntos computadora: {puntos_computadora}\n")
        elif computadora=="papel" and jugador == "piedra":
            puntos_computadora += 1
            print(f"Â¡Has perdido esta ronda!\n{nombre}: {jugador} ğŸ‘Š / Computadora: {computadora} ğŸ–\nPuntos jugador/a: {puntos_jugador} / Puntos computadora: {puntos_computadora}\n")
        elif computadora=="papel" and jugador == "tijera":
            puntos_jugador += 1
            print(f"Â¡Has ganado esta ronda!\n{nombre}: {jugador} âœŒ / Computadora: {computadora} ğŸ–\nPuntos jugador/a: {puntos_jugador} / Puntos computadora: {puntos_computadora}\n")
        # Mensaje de error si se introduce un opciÃ³n no vÃ¡lida
        else:
            print("Algo saliÃ³ mal... ğŸ˜• IntÃ©ntalo de nuevo.\n")

    # Comprobar si se han acumulado 3 puntos para ver quiÃ©n gana
    if puntos_jugador < 3 and puntos_computadora >= 3:
        print('Â¡VAYA, HAS PERDIDO EL JUEGO! ğŸ¥ˆ\n')
    else:
        print('Â¡ENHORABUENA, HAS GANADO EL JUEGO! ğŸ†\n')

    # Preguntar si se quiere jugar otra partida para salir del bucle o no
    jugar = input("Â¿Quieres jugar otra partida? s/n")

else:
    print('Gracias por jugar. Â¡HASTA PRONTO!')
           
```

## Ahorcado

Ahorcado: es un juego de adivinanzas en el que un jugador elige una palabra y el otro jugador debe adivinarla letra por letra. El jugador que adivina la palabra tiene un nÃºmero limitado de oportunidades para adivinar las letras que forman la palabra. Cada vez que el jugador adivina una letra incorrecta, se dibuja una parte del cuerpo del personaje del ahorcado en una horca. Si el jugador adivina todas las letras de la palabra antes de que se complete el dibujo del personaje del ahorcado, gana el juego. Si el dibujo del personaje del ahorcado se completa antes de adivinar la palabra, pierde el juego.

```python
# Juego 2. El Ahorcado

# Importamos el mÃ³dulo random para generar nÃºmeros aleatorios
import random

class Ahorcado:
    def __init__(self):
        # Inicializamos las vidas del jugador
        self.vidas = 6
        # Establecemos una lista de palabras de prueba, fÃ¡ciles de completar
        self.lista_palabras = ['hola', 'chao', 'casa']
        # Se elige una palabra al azar de la lista de palabras
        self.palabra = random.choice(self.lista_palabras)
        # Iniciamos la lista para almacenar las letras adivinadas por el jugador
        self.letras_adivinadas = ['_'] * len(self.palabra)
        # Creamos un conjunto de letras Ãºnicas para almacenar letras introducidas por el jugador
        self.letras_introducidas = set()  
        # Almacenamos el nombre del jugador
        self.nombre = None

    def intro(self):
        # Personalizamos el juego pidiendo el nombre
        self.nombre = input("Â¿CÃ³mo te llamas?").title()
        # Saludamos al jugador e introducimos las reglas del juego
        print(f"Â¿Preparado para jugar al Ahorcado, {self.nombre}?\n")
        print(f"Reglas:\n - Para salvarte de la horca, tienes que ir adivinando las letras que componen la palabra secreta.\n - Tienes 6 vidas. Â¡No te quedes colgado!\n")
        
        while True:
            # Llamamos al mÃ©todo jugar
            self.jugar()  
            # Al finalizar la partida, en este punto, se determina la continuaciÃ³n del juego
            if input("\nÂ¿Quieres jugar otra partida? (S/N)").lower() != 's':  # Verificamos si el jugador desea otra partida
                print('GRACIAS por jugar. Â¡Hasta pronto!')
                break
            else:
                # Reiniciamos las variables de instancia del constructor (excepto nombre, pues serÃ­a el mismo jugador)
                self.vidas = 6  # Reiniciamos las vidas
                self.palabra = random.choice(self.lista_palabras)
                self.letras_adivinadas = ['_'] * len(self.palabra)
                self.letras_introducidas = set()  # Reiniciamos las letras introducidas


    def jugar(self):

        # Iteramos sobre el bucle principal del juego
        while self.vidas > 0:
            # Mostramos el estado actual del ahorcado
            self.imprimir_ahorcado()
            letra = input('Ingresa una letra: \n').lower()
            # Verificamos si la entrada no es una letra
            if not letra.isalpha():  
                print("Por favor, introduce sÃ³lo LETRAS.")
                # Volvemos al inicio del bucle para pedir una nueva letra
                continue  
            try: 
                # Verificamos si la letra ya ha sido introducida previamente
                if letra in self.letras_introducidas:
                    print("Â¡CUIDADO! Ya lo intentaste con esa letra ğŸš«")
                    # Volvemos al inicio del bucle sin procesar la letra repetida
                    continue  
                # Intentamos adivinar la letra introducida
                self.adivinar_letra(letra)
                self.letras_introducidas.add(letra)
                # Verificamos si el jugador ha adivinado todas las letras de la palabra
                if '_' not in self.letras_adivinadas:
                    print('\nÂ¡Â¡YUHUUUUUU!!Â¡TE HAS SALVADO DE LA HORCA ğŸ‰!\n')
                    return  # Salimos del mÃ©todo si el jugador ha ganado
                # Verificamos si al jugador no le quedan vidas
                elif self.vidas == 0:
                    print('\nVAYA, HAS AGOTADO TODAS LAS VIDAS ğŸ’€')
                    print("""
                              ___
                             |   |
                            _O/  |
                             |   |
                            / \  |
                           ______|
                """)
                    print(f'La palabra era "{self.palabra}".\n')
                    return  # Salimos del mÃ©todo si el jugador ha perdido

            except ValueError:
                print("Algo saliÃ³ mal... ğŸ˜• IntÃ©ntalo de nuevo.\n")
        # Llamamos a reiniciar_partida solo si el jugador desea jugar otra partida
        self.reiniciar_partida()  

    def adivinar_letra(self, letra):
        if letra in self.palabra:
            for i in range(len(self.palabra)):
                if self.palabra[i] == letra:
                    self.letras_adivinadas[i] = letra
            print('\nÂ¡Correcto! ğŸ‘')
            # Verificamos que ya se hayan adivinado todas las letras de la palabra
            if '_' not in self.letras_adivinadas:
                print(f'La palabra es "{self.palabra}".')
            else: 
                print(''.join(self.letras_adivinadas))
        else:
            # Restamos una vida al jugador si la letra no estÃ¡ en la palabra
            self.vidas -= 1
            # Mostramos el nÃºmero de vidas restantes al jugador
            if self.vidas > 1:
                print(f'\nNo estÃ¡ la letra "{letra}". Te quedan {self.vidas} vidas.')
            elif self.vidas == 1:
                print(f'\nNo estÃ¡ la letra "{letra}". Te queda {self.vidas} vida.')


    def imprimir_ahorcado(self):

        if self.vidas == 6:
            print("""
                           ___
                          |   |
                              |
                              |
                              |
                        ______|
            """)

        if self.vidas == 5:
            print("""
                           ___
                          |   |
                          O   |
                              |
                              |
                        ______|
            """)
        elif self.vidas == 4:
            print("""
                           ___
                          |   |
                          O   |
                          |   |
                              |
                        ______|
            """)
        elif self.vidas == 3:
            print("""
                           ___
                          |   |
                         _O   |
                          |   |
                              |
                        ______|
            """)
        elif self.vidas == 2:
            print("""
                           ___
                          |   |
                         _O/  |
                          |   |
                              |
                        ______|
            """)
        elif self.vidas == 1:
            print("""
                           ___
                          |   |
                         _O/  |
                          |   |
                         /    |
                        ______|
            """)

    
    def reiniciar_partida(self):
        # MÃ©todo para preguntar al jugador si quiere jugar otra partida
        otra_partida = input("\nÂ¿Quieres jugar otra partida? (S/N)")
        if otra_partida.lower() == 's':
            # Reiniciamos las vidas, la palabra, las letras adivinadas y las letras introducidas para una nueva partida
            self.vidas = 6 
            self.palabra = random.choice(self.lista_palabras)
            self.letras_adivinadas = ['_'] * len(self.palabra)
            self.letras_introducidas = set()  
            return True
        elif otra_partida.lower() == 'n':
            print('Gracias por jugar. Â¡HASTA PRONTO!')
            return False


# Instanciamos y ejecutamos el juego
ahorcado = Ahorcado()
ahorcado.intro()

```

## Juego de preguntas y respuestas

Juego de preguntas y respuestas: es un juego en el que se hacen preguntas sobre una variedad de temas (en este caso, geografÃ­a) y los jugadores deben responder correctamente para avanzar. Las preguntas pueden ser de diferentes categorÃ­as, como historia, ciencias, entretenimiento o cultura general. El objetivo del juego es responder la mayor cantidad de preguntas correctamente para ganar el juego. Este juego puede ser jugado de forma individual o en equipo y es una forma divertida de aprender y poner a prueba tus conocimientos.

### CÃ³digo

```python
# Juego 3. Juego de Preguntas y Respuestas

def imprimir_separador(caracter, longitud):
    print(caracter * longitud)

def jugar():
    
    saludo = input("Â¡Hola! Â¿CÃ³mo te llamas? ")
    print("\n\nÂ¡Bienvenid@ a PREGUNTAS Y RESPUESTAS,",saludo.title(), "!. Â¿Te atreves con el bloque de GEOGRAFÃA?")
    imprimir_separador("~~", 42)
    print('\n')

    # Defino diccionario de preguntas y respuestas
    dic_geografia = {
        "1. Â¿CuÃ¡l es el rÃ­o mÃ¡s largo de la PenÃ­nsula IbÃ©rica?" : "tajo",
        "2. Â¿CuÃ¡l es la capital de BÃ©lgica?" : "bruselas",
        "3. Â¿QuÃ© paÃ­s tiene mÃ¡s habitantes?" : "india",
        "4. Â¿QuÃ© paÃ­s es el mÃ¡s grande del mundo?" : "rusia",
        "5. Â¿En quÃ© paÃ­s estÃ¡ la catarata mÃ¡s alta del mundo?" : "venezuela",
        "6. Â¿CuÃ¡l es la montaÃ±a mÃ¡s alta del mundo?" : "everest",
        "7. Â¿CÃ³mo se llaman la cordillera entre Argentina y Chile?" : "andes",
    }


    # Defino contador de preguntas acertadas y preguntas fallidas
    preguntas_acertadas = 0
    preguntas_fallidas = 0

    pregunta_actual = iter(dic_geografia.items())

    while preguntas_acertadas < 5 and preguntas_fallidas < 3:
        # Usamos next, funciÃ³n que permite obtener le prÃ³ximo elemento de un iterador
        pregunta, respuesta = next(pregunta_actual)
        print(pregunta)
        respuesta_user = input(pregunta)
        if respuesta in respuesta_user.lower():
            preguntas_acertadas += 1
            print(f"Correcto ğŸ‘, llevas {preguntas_acertadas} preguntas acertadas y {preguntas_fallidas} fallidas!")
            imprimir_separador('-', 20)
            print('\n')
        else:
            preguntas_fallidas += 1
            print(f"No has acertado ğŸ˜£... Llevas {preguntas_acertadas} preguntas acertadas y {preguntas_fallidas} fallidas.")
            imprimir_separador('-', 20)
            print('\n')

    if preguntas_acertadas >= 5 and preguntas_fallidas < 3:
        print(f"ERES UN AUTÃ‰NTICO EXPLORADOR ğŸ§­. Â¡Has acertado {preguntas_acertadas} preguntas y fallado solo {preguntas_fallidas}!\n")
    else:
        print(f"Sigue estudiando ğŸ“–... pronto lo conseguirÃ¡s. Has acertado {preguntas_acertadas} preguntas y fallado {preguntas_fallidas}.\n")

    otra_partida = input("\nÂ¿Quieres jugar otra vez? (S/N) ")
    if otra_partida.lower() == 's':
        # Si el jugador desea jugar otra partida, reiniciamos los contadores y volvemos a empezar
        jugar()
    else:
        # Si el jugador no desea jugar mÃ¡s, mostramos el mensaje de despedida
        print('\nGracias por jugar. Â¡HASTA PRONTO!')


# Llamamos al mÃ©todo para iniciar el juego
jugar()
```
